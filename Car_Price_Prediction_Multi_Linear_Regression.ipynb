```python
cols_to_cont_less_than_positive_threshold = np.where(df.corr()['price']<=0.5)
cols_to_cont_more_than_negative_threshold = np.where(df.corr()['price']>= -0.5)
#cols_to_cont = cols_to_cont.where(cols_to_cont == True)
cols_to_cont_less_than_positive_threshold
cols_to_cont_more_than_negative_threshold
cols_to_drop = np.intersect1d(cols_to_cont_less_than_positive_threshold, cols_to_cont_more_than_negative_threshold)
cols_to_drop.shape
```

```python
c = [ i for i in range(1, 206, 1)]
fig = plt.figure()

plt.plot(c, y, color='blue', linewidth=2.5, linestyle='-')

plt.plot(c, pred, color='red', linewidth=2.5, linestyle='-')

fig.suptitle('Actual and Predicted', fontsize=20)

plt.xlabel('Index', fontsize=18)

plt.ylabel('Car Price', fontsize=16)
```

```python
c = [ i for i in range(1, 206, 1)]
plt.figure()
plt.plot(c, y-pred, color='blue', linewidth=2.5, linestyle='-')
plt.suptitle('Error Graph', fontsize=20)
plt.xlabel('Index', fontsize=18)
plt.ylabel('Car Price', fontsize=16)

from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y, pred)
r2 = r2_score(y, pred)

print(f'The Mean Squared Error is {mse}')
print(f'The R Squared Score is {r2}')
```
